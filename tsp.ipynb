```python
n = 4
graph= [[ 0, 10, 15, 20 ], 
        [ 10, 0, 35, 25 ], 
        [ 15, 35, 0, 30 ], 
        [ 20, 25, 30, 0 ]] 
answer = [] 
def tsp(v, currPos, count, cost): 
    global graph
    global n
  
    if (count == n and graph[currPos][0]): 
        answer.append(cost + graph[currPos][0]) 
        return
 
    for i in range(n): 
        if (v[i] == False and graph[currPos][i]): 
              
            # Mark as visited 
            v[i] = True
            tsp(v, i, count + 1,  
                cost + graph[currPos][i]) 
              
            v[i] = False
            
def newGraph(newNode):
    global graph;
    graph.append(newNode)
    
    temp_graph = []
    
    for i in range(n):
        temp_row = graph[i]
        if newNode[i] != 0:
            temp_row.append(int(newNode[i]))
            temp_graph.append(temp_row)
        else:
            temp_row.append(0)
            temp_graph.append(temp_row)

    graph = temp_graph
    
    
def addNode():
    global n
    n = n+1
    
    newNode = []
    
    for i in range(n):
        temp = input("Distance from node {}:".format(i+1))
        newNode.append(int(temp))
        
    newGraph(newNode)
    
def newGraphDelete(nodeToBeDeleted):
    global graph;
    
    new_graph = []
    
    temp_graph = graph
    
    n_delete = nodeToBeDeleted - 1
    
    del temp_graph[n_delete]
    
    for i in range(len(temp_graph)):
        temp_row = temp_graph[i]
        del temp_row[n_delete]
        new_graph.append(temp_row)
        
    graph = new_graph
    print(graph)
    
def deleteNode():
    global n
    n = n-1
    nodeToBeDeleted = int(input("Enter the node you want to delete: "))
    
    newGraphDelete(nodeToBeDeleted)
     
    
    
def baseSetup():

    v = [False for i in range(n)] 
      
    v[0] = True
  
    tsp(v, 0, 1, 0) 
    
  
    print(min(answer))
 
  
if __name__ == '__main__': 
    
    print("The initial graph tsp answer is ")
    baseSetup()
    
    command = input('You want to add (or a) or delete (or d) or exit (or e): ')
    command = command.lower()
    if command == "add" or command == "a" :
        addNode()
        baseSetup()
    if command == "delete" or command == "d":
        deleteNode()
        baseSetup()
    if command == "exit" or command == "e":
        print("Hope you found the tool useful. Thank you for using the tool")
        exit()
```
