```python
# n: number of nodes. global variable and will change when user add/delee the node.
# Inititally there are 4 nodes given in the program statement.
n = 4 

# creating a distance graph based on the program statement.
#global variable and will change when user add/delee the node.
graph= [[ 0, 10, 15, 20 ], 
        [ 10, 0, 35, 25 ], 
        [ 15, 35, 0, 30 ], 
        [ 20, 25, 30, 0 ]] 
answer = [] 

#Calculating the shortest route in TSP
def tsp(v, currPos, count, cost): 
    global graph
    global n
    global answer
  
    if (count == n and graph[currPos][0]): 
        answer.append(cost + graph[currPos][0]) 
        return
 
    for i in range(n): 
        if (v[i] == False and graph[currPos][i]): 
              
            # Mark as visited 
            v[i] = True
            tsp(v, i, count + 1,  
                cost + graph[currPos][i]) 
              
            v[i] = False
            

# this function re-arranges the whole graph whenever user adds a new node.            
def newGraph(newNode):
    global graph;
    graph.append(newNode)
    
    temp_graph = []
    
    for i in range(n):
        temp_row = graph[i]
        if newNode[i] != 0:
            temp_row.append(int(newNode[i]))
            temp_graph.append(temp_row)
        else:
            temp_row.append(0)
            temp_graph.append(temp_row)

    graph = temp_graph
    

# this function accepts new node from the user asking the distance.
# new node is entered in the form of a list but not as a whole distance matrix.
def addNode():
    global n
    n = n+1
    
    newNode = []
    
    for i in range(n):
        temp = input("Distance from node {}:".format(i+1))
        newNode.append(int(temp))
        
    newGraph(newNode)
    
# this function re-arranges the whole graph whenever user delets a new node.
# takes the the node number that has to be deleted.
def newGraphDelete(nodeToBeDeleted):
    global graph;
    
    new_graph = []
    
    temp_graph = graph
    
    n_delete = nodeToBeDeleted - 1
    
    del temp_graph[n_delete]
    
    for i in range(len(temp_graph)):
        temp_row = temp_graph[i]
        del temp_row[n_delete]
        new_graph.append(temp_row)
        
    graph = new_graph
    print(graph)

# function to get the number of the node that has to be deleted.
def deleteNode():
    global n
    n = n-1
    nodeToBeDeleted = int(input("Enter the node you want to delete: "))
    
    newGraphDelete(nodeToBeDeleted)
     
    
# This function is called always whenever the user adds/delets the node. 
def baseSetup():
    
    global answer

    v = [False for i in range(n)] 
      
    v[0] = True
  
    tsp(v, 0, 1, 0) 
    
  
    print(min(answer))
 
  
if __name__ == '__main__': 
    
    print("The initial graph tsp answer is ")
    baseSetup()
    
    boolean = True
    
    while boolean:
        command = input('You want to add (or a) or delete (or d) or exit (or e): ')
        command = command.lower()
        if command == "add" or command == "a" :
            addNode()
            baseSetup()
        if command == "delete" or command == "d":
            deleteNode()
            baseSetup()
        if command == "exit" or command == "e":
            print("Hope you found the tool useful. Thank you for using the tool")
            boolean = False
            
    exit()
```
